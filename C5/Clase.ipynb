## Pandas

Es una librería que agrega funcionalidades sobre Numpy. Principalmente combina la eficiencia en los cálculos numérico y el manejo de arrays con el manejo de diversos tipos de datos que ayudan a hacer la limpieza de datos.

Además incluye un manejo más intuitivo y con etiquetas de la selección de datos.

## Formas de filtrar datos especificas de pandas: Index

Pandas permite elegir los datos para su manipulación de una manera muy intuitiva y mediante etiquetas. Vamos a ver diferentes maneras de usar esas herramientas. Tiene propiedades de conjuntos y es no mutable.
